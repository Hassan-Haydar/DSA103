```python
import pandas as pd  # noqa: I001
from sklearn.datasets import load_wine
import seaborn as sns
import matplotlib.pyplot as plt

wine = load_wine()

df = pd.DataFrame(wine.data, columns=wine.feature_names)



#df.columns

df["target"] = wine.target

df.head()

#Print the shape, column names, and first 5 rows.
print("Shape:", df.shape)

print(f"\nColumns: {df.columns}")   # all column names
print("\nFirst 5 rows:\n", df.head()) #head takes the first 5 rows

```

```python
df.describe()
```

```python
#Count how many samples belong to each class (value_counts()).
print("\nClass distribution:\n", df['target'].value_counts())
#Check for missing values (df.isna().sum()).
print("\nMissing values:\n", df.isna().sum())
```

```python
#Compute the correlation matrix and display it as a heatmap (seaborn.heatmap).

corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
